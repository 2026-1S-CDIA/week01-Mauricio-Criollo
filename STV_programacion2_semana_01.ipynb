# 1. ¿Qué es la Programación Orientada a Objetos?

La **Programación Orientada a Objetos (POO)** es un paradigma de programación que organiza el código alrededor de **objetos**.

Un objeto representa una entidad del mundo real o conceptual, por ejemplo:

- un estudiante
- un vehículo
- una cuenta bancaria
- un producto

Cada objeto tiene:

- **atributos**: características o datos
- **métodos**: acciones o comportamientos

## Ejemplos del mundo real
- Un **estudiante** tiene nombre, edad y carrera.
- Un **estudiante** puede estudiar, presentarse o saludar.
- Un **vehículo** tiene marca, modelo y color.
- Un **vehículo** puede arrancar, frenar y girar.

La POO permite modelar estos elementos de forma más organizada y reutilizable.

# 2. Clase y objeto

## Clase
Una **clase** es una plantilla o molde que define cómo serán los objetos.

## Objeto
Un **objeto** es una instancia concreta de una clase.

### Ejemplo conceptual
- Clase: `Estudiante`
- Objetos:
  - Juan
  - María
  - Pedro

Todos ellos pertenecen a la clase `Estudiante`, pero cada uno puede tener valores diferentes en sus atributos.

```python
# Primer ejemplo simple de clase

class Estudiante:
    pass

# Crear objetos a partir de la clase
est1 = Estudiante()
est2 = Estudiante()

print(type(est1))
print(type(est2))
```

En el ejemplo anterior:

- `class Estudiante:` define una clase
- `pass` indica que, por ahora, la clase no tiene contenido
- `est1` y `est2` son objetos creados a partir de la clase `Estudiante`

Aunque ambos son de la misma clase, representan instancias diferentes.

# 3. Atributos

Los **atributos** son variables asociadas a un objeto.  
Sirven para almacenar información o características.

## Ejemplos de atributos en un estudiante
- nombre
- edad
- carrera
- semestre

```python
class Estudiante:
    def __init__(self, nombre, edad, carrera):
        self.nombre = nombre
        self.edad = edad
        self.carrera = carrera

# Crear objetos
est1 = Estudiante("Ana", 20, "Ingeniería en TI")
est2 = Estudiante("Luis", 22, "Pedagogía de la Informática")

print(est1.nombre)
print(est1.edad)
print(est1.carrera)

print(est2.nombre)
print(est2.edad)
print(est2.carrera)
```

## Explicación del método `__init__`

El método `__init__` es un **constructor**.  
Se ejecuta automáticamente cuando se crea un objeto.

### Partes importantes
- `self`: representa al objeto actual
- `self.nombre = nombre`: guarda el valor recibido en el atributo del objeto
- `self.edad = edad`: asigna la edad al objeto
- `self.carrera = carrera`: asigna la carrera al objeto

Cada objeto tendrá sus propios valores.

# 4. Métodos

Los **métodos** son funciones definidas dentro de una clase.  
Describen las acciones o comportamientos que puede realizar un objeto.

```python
class Estudiante:
    def __init__(self, nombre, edad, carrera):
        self.nombre = nombre
        self.edad = edad
        self.carrera = carrera

    def saludar(self):
        print(f"Hola, mi nombre es {self.nombre} y estudio {self.carrera}.")

    def cumplir_anios(self):
        self.edad += 1
        print(f"Ahora tengo {self.edad} años.")

# Crear objeto
est1 = Estudiante("Carla", 19, "Software")

# Usar métodos
est1.saludar()
est1.cumplir_anios()
```

En este ejemplo:

- `saludar()` muestra información del objeto
- `cumplir_anios()` modifica el valor del atributo `edad`

Esto demuestra que los objetos no solo almacenan datos, sino que también pueden realizar acciones.

# 5. Sintaxis básica de una clase en Python

La estructura general es la siguiente:

```python
class NombreDeLaClase:
    def __init__(self, parametro1, parametro2):
        self.atributo1 = parametro1
        self.atributo2 = parametro2

    def metodo1(self):
        print("Ejecutando método 1")

    def metodo2(self):
        print("Ejecutando método 2")
```

## Elementos importantes de la sintaxis

- `class` se usa para definir una clase
- `__init__` inicializa los atributos del objeto
- `self` hace referencia al propio objeto
- Los métodos siempre reciben `self` como primer parámetro
- La indentación es obligatoria en Python

# 6. Ejemplo guiado: clase Vehiculo

Ahora construiremos otra clase para reforzar los conceptos.

```python
class Vehiculo:
    def __init__(self, marca, modelo, color):
        self.marca = marca
        self.modelo = modelo
        self.color = color

    def mostrar_info(self):
        print(f"Vehículo: {self.marca} {self.modelo}, color {self.color}")

    def arrancar(self):
        print(f"El vehículo {self.marca} {self.modelo} ha arrancado.")

# Crear objetos
v1 = Vehiculo("Toyota", "Corolla", "Blanco")
v2 = Vehiculo("Chevrolet", "Spark", "Rojo")

v1.mostrar_info()
v1.arrancar()

v2.mostrar_info()
v2.arrancar()
```

# 7. Diferencia entre clase y objeto

## Clase
Es la definición general.

## Objeto
Es una instancia concreta con datos específicos.

### Ejemplo
- Clase: `Vehiculo`
- Objeto 1: Toyota Corolla blanco
- Objeto 2: Chevrolet Spark rojo

Ambos objetos pertenecen a la misma clase, pero tienen diferentes valores en sus atributos.
